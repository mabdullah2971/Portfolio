```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('SalarySurvey2021_updated.csv')
```

# Which industries pay the most?

```python
# Groupby industry and calculate average annual salary
avg_salaries = df.groupby('Industry')['Annual_Salary'].mean().sort_values(ascending=True)

# Create a bar chart
avg_salaries.plot(kind='barh')

plt.title('Average Annual Salary by Industry')
plt.xlabel('Average Annual Salary (£)')
plt.ylabel('')

plt.show()
```
